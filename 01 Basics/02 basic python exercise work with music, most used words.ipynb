```python
from os import listdir
from os.path import isfile, join
import zipfile
import requests
import re

html = {}

singers= ['eminem','madonna']
for i in singers:
    files = listdir('songtexte/'+i)
    
    for file in files[:3]:
       
        print(file)
        out = open('songtexte/'+i+'/'+file)
        html.update({file:out.read()})
        
print(html)
```

```python
lyrics = {}
for i in html:
    pattern = '<!-- Usage of azlyrics.com .* agreement. Sorry about that. -->([\w\W]+)<!-- MxM banner -->'
    text = re.findall(pattern,html[i])
    lyrics.update({i:text[0].replace('<br>','').replace('</div>','').replace('<i>','').replace('</i>','')}) 
    
```

```python
for i in lyrics:
    print(i)
    file = open('cleaned/'+i,'w')  
    file.write(lyrics[i]) 
    file.close()
```

```python
from collections import Counter
words = []
for i in lyrics:
    
    words += i.split()

c = Counter(words)
c.most_common(50)
```
