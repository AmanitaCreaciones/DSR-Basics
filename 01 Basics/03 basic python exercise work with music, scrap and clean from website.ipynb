# Basic python exercise, scrap and clean

```python
import requests
url = "http://www.azlyrics.com/r/rogerwaters.html#46655"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
r = requests.get(url, headers=headers)
print(r.text)
```

## we scrap the songs link

```python
import re
songs = []
# buscamos lo que cumpla este patrón, y guardamos lo que hay entre paréntesis en una lista de (en este caso), longitud 2
pattern = '<a href="..(/lyrics.*)" target="_blank">(.*)</a>'
links = re.findall(pattern,r.text)
for i in links:
    link= 'http://www.azlyrics.com'+i[0]
    songs.append(link)
    

print(songs)
```

```python
## we scrap the lyrics of the previous links with 7 sec delay between them
```

```python
import time
html_data = []
#we only load 3 times if each singer, so we don't get kicked out of the website
for i in songs[:3]:
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    r = requests.get(i, headers=headers)
    html_data.append(r.text)
    # having 7 secconds delay in each loop heps us not to have the ip banned from the website
    time.sleep(7)
print(html_data)   
```

# clean the data from the html tags

```python
for i in html_data:
    pattern = '<!-- Usage of azlyrics.com .* agreement. Sorry about that. -->([\w\W]+)<!-- MxM banner -->'
    text = re.findall(pattern,i)
    text[0] = text[0].replace('<br>','').replace('</div>','')
    print(text[0])
    
```
